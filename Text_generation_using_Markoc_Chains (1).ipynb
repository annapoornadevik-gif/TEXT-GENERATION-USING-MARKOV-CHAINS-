```python
# ============================================================
# TEXT GENERATION USING MARKOV CHAINS
# ============================================================

# Step 1: Import Libraries
import random
import re
from collections import defaultdict

# ============================================================
# Step 2: Sample Training Text
# ============================================================

text = """
Artificial Intelligence is transforming the world.
Machine learning is a branch of artificial intelligence.
Deep learning is a subset of machine learning.
Artificial intelligence can automate tasks.
Machine learning models learn from data.
Data science and artificial intelligence are related fields.
"""

# ============================================================
# Step 3: Text Preprocessing
# ============================================================

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    # Split into words
    words = text.split()

    return words

words = preprocess_text(text)

print("Processed Words:")
print(words)

# ============================================================
# Step 4: Build Markov Chain Model
# ============================================================

def build_markov_chain(words, n=1):
    """
    n = order of Markov chain
    n=1 -> unigram
    n=2 -> bigram
    """

    markov_chain = defaultdict(list)

    for i in range(len(words) - n):
        key = tuple(words[i:i+n])
        next_word = words[i+n]

        markov_chain[key].append(next_word)

    return markov_chain

# Build Bigram Model
n = 2
markov_chain = build_markov_chain(words, n)

print("\nMarkov Chain Sample:")
for key, value in list(markov_chain.items())[:5]:
    print(f"{key} -> {value}")

# ============================================================
# Step 5: Generate Text
# ============================================================

def generate_text(markov_chain, length=20):

    # Random starting state
    current_state = random.choice(list(markov_chain.keys()))

    generated_words = list(current_state)

    for _ in range(length):

        current_key = tuple(generated_words[-n:])

        possible_next_words = markov_chain.get(current_key)

        if not possible_next_words:
            break

        next_word = random.choice(possible_next_words)

        generated_words.append(next_word)

    return ' '.join(generated_words)

# ============================================================
# Step 6: Generate Output
# ============================================================

generated_text = generate_text(markov_chain, length=30)

print("\nGenerated Text:\n")
print(generated_text)

# ============================================================
# Step 7: Generate Multiple Sentences
# ============================================================

print("\nMultiple Generated Sentences:\n")

for i in range(5):
    print(f"{i+1}. {generate_text(markov_chain, length=20)}")
```
